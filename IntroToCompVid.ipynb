### Resize image

````python
# Resize image
import cv2 
from matplotlib import pyplot as plt

img = cv2.imread('furina.jpg') 

# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Get the original dimensions
original_height, original_width, channels = img.shape

# Define the scale factor for resizing
skala = 0.25

# Calculate the new dimensions
new_height = int(original_height * skala)
new_width = int(original_width * skala)


# Resize the image
resized_image = cv2.resize(img, (new_width, new_height))
print(f"Height: {new_height}, Width: {new_width}, Channels: {channels}")
# Show image using Matplotlib as cv2.imshow will cause jupyter kernel crashed.
plt.imshow(resized_image)
````

## Color Detection

````python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('red.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# lower range of red color in HSV
lower_range = (0, 50, 50)
# upper range of red color in HSV
upper_range = (150, 255, 255)

mask = cv2.inRange(hsv_img, lower_range, upper_range)

color_image = cv2.bitwise_and(img, img, mask=mask)

# Display the color of the image
plt.imshow(color_image)
````

````python
color_image = cv2.cvtColor(color_image, cv2.COLOR_RGB2BGR)
cv2.imshow('masking',color_image)
cv2.waitKey(0) 
cv2.destroyAllWindows()
````

### Membuat Contours

````python
import cv2

cap = cv2.VideoCapture(0)

while True:
  success, img = cap.read()
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  lower_range = (0, 50, 50)
  upper_range = (150, 255, 255)
  mask = cv2.inRange(hsv_img, lower_range, upper_range)
  color_image = cv2.bitwise_and(img, img, mask=mask)
  color_image = cv2.cvtColor(color_image, cv2.COLOR_RGB2BGR)

  contours, hierarchy = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  # center = None
  
  if len(contours) > 0:
    c = max(contours, key=cv2.contourArea) # cari luasan contour 
    rect = cv2.minAreaRect(c) # bikin persegi berdasarkan minimal luasan contour
    ((x,y), (width, height), rotation) = rect
    
    x1 = x-width/2
    y1 = y-height/2

    x2 = x+width/2
    y2 = y+height/2

    cv2.rectangle(img, (int(x1),int(y1)),(int(x2),int(y2)), (255, 0,0), 3)
    cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
    cv2.circle(img, (int(x),int(y)), 3, (0, 0, 255), -1)
    cv2.putText(img, 'Detected', (int(x1),int(y1)), cv2.FONT_ITALIC, 4, (255, 0, 0), 3)

  img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
  cv2.imshow('red', img)
  if cv2.waitKey(1) & 0xFF == ord('q'):
    break

cap.release()
cv2.destroyAllWindows()
````

## Contour di gambar

````python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('red.jpg')

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lower_range = (0, 50, 50)
upper_range = (150, 255, 255)
mask = cv2.inRange(hsv_img, lower_range, upper_range)
color_image = cv2.bitwise_and(img, img, mask=mask)
color_image = cv2.cvtColor(color_image, cv2.COLOR_RGB2BGR)

contours, hierarchy = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# center = None

if len(contours) > 0:
  c = max(contours, key=cv2.contourArea) # cari luasan contour 
  rect = cv2.minAreaRect(c) # bikin persegi berdasarkan minimal luasan contour
  ((x,y), (width, height), rotation) = rect
  
  x1 = x-width/2
  y1 = y-height/2

  x2 = x+width/2
  y2 = y+height/2

  cv2.rectangle(img, (int(x1),int(y1)),(int(x2),int(y2)), (255, 0,0), 3)
  cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
  

plt.imshow(img)
````
