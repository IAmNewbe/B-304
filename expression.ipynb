<h1>Ekspresi</h1>
Ekspresi pada pemrograman merupakan kombinasi dari satu atau lebih variabel, konstanta, operator, dan fungsi yang bermakna untuk menghasilkan suatu nilai dalam suatu tipe tertentu.

```python
angka = [2, 4, 6, 8]
huruf = ['P', 'Y', 'T', 'H', 'O', 'N']
lagi = ['1','halo']
gabung = angka + huruf + lagi #ekspresi

print(gabung)
```

```python
learn = ['P', 'Y', 'T', 'H', 'O', 'N']
bil = [1, 2, 3, 4, 5, 6, 7, 8]
replikasi = learn * 2
replikasi_bil = bil * 2

print(replikasi)
print(replikasi_bil)
```

<h2>Else setelah For</h2>
Pada Python juga dikenal else setelah for yang berfungsi untuk perulangan bersifat pencarian. Else setelah for ini bisa dikatakan sebagai memberikan jalan keluar program saat pencarian tidak ditemukan.

```python
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    if num == 4:
        print("Angka ditemukan! Program berhenti!")
        break
else:
    print("Angka tidak ditemukan.")
```

<h2>Else setelah While</h2>
Berbeda dengan else setelah for, pada statement else setelah while, blok statement else akan selalu dieksekusi saat kondisi pada while menjadi salah. 

```python
count = 0

while count < 3:
    print("Dicoding Indonesia")
    count += 1
else:
    print("Blok else dieksekusi karena kondisi pada while salah (3<3 == False).")
```

<h2>Continue</h2>
pernyataan untuk membuat iterasi berhenti, kemudian melanjutkan ke iterasi berikutnya. Continue seolah mengabaikan pernyataan (statement) yang berada antara continue hingga akhir blok.

```python
for huruf in 'Dico ding':
    if huruf == ' ':
        continue
    print('Huruf saat ini: {}'.format(huruf))
```

<h2>Arrays</h2>

```python
var_arr = [1, 2, 3, 4, 5]

for i in range(len(var_arr)):
    current_element = var_arr[i]
    next_index = i+1
    
    if next_index < len(var_arr):
        next_element = var_arr[next_index]
    else:
        next_element = None
        
    print(f"Current element: {current_element}, next elements: {next_element}")
```

<h2>Operasi Matriks pada Python</h2>

```python
# Membuat matriks 2x2
var_mat = [[5, 0],
          [1, -2]]
def_mat = [[0 for j in range(2)] for i in range(2)]

for i in range(len(var_mat)):
  for j in range(len(var_mat[0])):
    def_mat[i][j] = var_mat[i][j]*2

print(def_mat)
```

```python
import numpy as np

var_mat = np.array([[5, 0],
                    [1, -2]])

result = var_mat * 2
print(result)
```
